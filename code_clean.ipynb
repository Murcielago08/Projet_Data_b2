```python
import pandas as pd
import os
import string
```

```python
df_accidents = pd.read_csv("accidentsVelo.csv")
df_accidents.columns
df_accidents.info()

nom_fichier = "accidentsVelo_clean.csv"

if os.path.exists(nom_fichier):
    os.remove(nom_fichier)
    print("Le fichier a été supprimé.")
else:
    print("Le fichier n'existe pas.")

```

```python
df_accidents.loc[df_accidents['Num_Acc'] == 36]
```

```python
df_accidents = df_accidents[df_accidents["Num_Acc"] != 36]
df_accidents.loc[df_accidents['Num_Acc'] == 36]
```

```python
df_accidents["mois"] = df_accidents["mois"].str.replace("é","e").str.replace("û","u")
df_accidents.loc[df_accidents['mois'] == 'aout']
```

```python
df_accidents.loc[df_accidents['mois'] == 'fevrier']
```

```python
df_accidents["lat"] = df_accidents["lat"].astype(str).apply(lambda x: x.replace(',', '.'))
df_accidents["lat"]
```

```python
df_accidents["long"] = df_accidents["long"].astype(str).apply(lambda x: x.replace(',', '.'))
df_accidents["long"]
```

```python
def different_de_zero_apres_virgule(nombre, precision=1e-10):
    return abs(nombre - int(nombre)) > precision

def au_moins_un_different_de_zero_apres_virgule(colonne):
    return colonne.apply(different_de_zero_apres_virgule).any()


resultat_final = au_moins_un_different_de_zero_apres_virgule(df_accidents['agg'])
print(resultat_final)
```

```python
def modifier_heure_minute(heure_minute):
    time = heure_minute.replace(':', "")
    zeros_to_add = 4 - len(time)
    time_correct_length = "0" * zeros_to_add + time
    formatted_time = time_correct_length[:2] + ":" + time_correct_length[2:]
    return formatted_time

df_accidents["hrmn"] = df_accidents["hrmn"].apply(modifier_heure_minute)
df_accidents["hrmn"]
```

```python
df_accidents["agg"] = df_accidents["agg"].replace(1,"Hors-Agglomeration")
df_accidents["agg"] = df_accidents["agg"].replace(2,"Agglomeration")
df_accidents["agg"]
```

```python
df_accidents["age"] = pd.to_numeric(df_accidents["age"], errors='coerce').astype('Int64')
df_accidents["age"]
```

```python
df_accidents["col"] = df_accidents["col"].replace(1,"Deux vehicules - frontale")
df_accidents["col"] = df_accidents["col"].replace(2,"Deux vehicules - par l'arrière")
df_accidents["col"] = df_accidents["col"].replace(3,"Deux vehicules - par le cote")
df_accidents["col"] = df_accidents["col"].replace(4,"Trois vehicules et plus - en chaine")
df_accidents["col"] = df_accidents["col"].replace(5,"Trois vehicules et plus - collisions multiples")
df_accidents["col"] = df_accidents["col"].replace(6,"Autre")
df_accidents["col"] = df_accidents["col"].replace(7,"Sans collision")
df_accidents["col"]
```

```python
df_accidents["lum"] = df_accidents["lum"].replace(1,"Plein jour")
df_accidents["lum"] = df_accidents["lum"].replace(2,"Crepuscule")
df_accidents["lum"] = df_accidents["lum"].replace(3,"Nuit sans eclairage public")
df_accidents["lum"] = df_accidents["lum"].replace(4,"Nuit avec eclairage public non allume")
df_accidents["lum"] = df_accidents["lum"].replace(5,"Eclairage de nuit")
df_accidents["lum"]
```

```python
df_accidents["atm"] = df_accidents["atm"].replace(1,"Autre")
df_accidents["atm"] = df_accidents["atm"].replace(2,"Pluie legere")
df_accidents["atm"] = df_accidents["atm"].replace(3,"Pluie forte")
df_accidents["atm"] = df_accidents["atm"].replace(4,"Brouillard - fumee")
df_accidents["atm"] = df_accidents["atm"].replace(5,"Vent fort - tempete")
df_accidents["atm"] = df_accidents["atm"].replace(6,"Grele")
df_accidents["atm"] = df_accidents["atm"].replace(7,"Plein jour")
df_accidents["atm"] = df_accidents["atm"].replace(8,"Temps couvert")
df_accidents["atm"] = df_accidents["atm"].replace(9,"Non repertorier")
df_accidents["atm"]
```

```python
df_accidents["grav"] = df_accidents["grav"].replace(1,"Indemne")
df_accidents["grav"] = df_accidents["grav"].replace(2,"Mort")
df_accidents["grav"] = df_accidents["grav"].replace(3,"Hospitalise")
df_accidents["grav"] = df_accidents["grav"].replace(4,"Leger")
df_accidents["grav"]
```

```python
df_accidents["sexe"] = df_accidents["sexe"].replace(1,"Masculin")
df_accidents["sexe"] = df_accidents["sexe"].replace(2,"Feminin")
df_accidents["sexe"]
```

```python
df_accidents["trajet"] = df_accidents["trajet"].replace(-1,None)
df_accidents["trajet"] = df_accidents["trajet"].replace(0,None)
df_accidents["trajet"] = df_accidents["trajet"].replace(1,"Domicile – travail")
df_accidents["trajet"] = df_accidents["trajet"].replace(2,"Domicile - ecole")
df_accidents["trajet"] = df_accidents["trajet"].replace(3,"Courses - achats")
df_accidents["trajet"] = df_accidents["trajet"].replace(4,"Utilisation professionnelle")
df_accidents["trajet"] = df_accidents["trajet"].replace(5,"Loisir promenade")
df_accidents["trajet"] = df_accidents["trajet"].replace(6,None)
df_accidents["trajet"] = df_accidents["trajet"].replace(7,None)
df_accidents["trajet"] = df_accidents["trajet"].replace(8,None)
df_accidents["trajet"] = df_accidents["trajet"].replace(9,"Autre")
df_accidents["trajet"] = df_accidents["trajet"].replace("",None)
df_accidents["trajet"]
```

```python
df_accidents["obs"] = pd.to_numeric(df_accidents["obs"], errors='coerce').astype('Int64')
df_accidents["obs"]
```

```python
df_accidents["obsm"] = pd.to_numeric(df_accidents["obsm"], errors='coerce').astype('Int64')
df_accidents["obsm"]
```

```python
df_accidents["manv"] = pd.to_numeric(df_accidents["manv"], errors='coerce').astype('Int64')
df_accidents["manv"]
```

```python
df_accidents["numVehicules"] = pd.to_numeric(df_accidents["numVehicules"], errors='coerce').astype('Int64')
df_accidents["numVehicules"]
```
